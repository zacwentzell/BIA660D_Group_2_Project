```python
## installation and more details in https://github.com/csurfer/rake-nltk
##https://www.airpair.com/nlp/keyword-extraction-tutorial#3-2-applying-rake-on-a-piece-of-text
from rake_nltk import Rake
import pandas as pd
import numpy as np
```

```python
r = Rake()
```

```python
text ='''Solid product Works as well as flashlights 3 or 4 times the cost of this one, at least. I will buy another pair, soon.'''
```

```python
type(text)
```

```python
r.extract_keywords_from_text(text)
```

```python
r.get_ranked_phrases()
```
